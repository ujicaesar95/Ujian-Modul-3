## 9. Use K Fold (K=5,shuffle=True,random_state=42), every loop use RandomOverSampler for the training data, and then use GradientBoostingClassifier for the model

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score, matthews_corrcoef, f1_score
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split, KFold, cross_validate
from sklearn.ensemble import GradientBoostingClassifier
```

```python
from sklearn.model_selection import train_test_split


data = dfpca
target = df['Class']

X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.3, random_state=101)
```

```python
import imblearn
from imblearn.over_sampling import RandomOverSampler

ros = RandomOverSampler(random_state=101)
X_ros, y_ros = ros.fit_sample(X_train,y_train)
pd.Series(y_ros).value_counts()
```

```python
X_ros
```

```python
def calc_train_error(X_ros, y_ros, model):
    predictions = model.predict(X_ros)
    predictProba = model.predict_proba(X_ros)
    matt = matthews_corrcoef(y_ros, predictions)
    f1 = f1_score(y_ros, predictions, average='macro')
    report = classification_report(y_ros, predictions)
    roc_auc = roc_auc_score(y_ros, predictProba[:,1])
    accuracy = accuracy_score(y_ros, predictions)
    confMatrix = confusion_matrix(y_ros, predictions)
    return {
        'report' : report,
        'matthews' : matt,
        'f1' : f1,
        'roc' : roc_auc,
        'accuracy' : accuracy,
        'confusion' : confMatrix
    }

def calc_validation_error(X_test, y_test, model):
    predictions = model.predict(X_test)
    predictProba = model.predict_proba(X_test)
    matt = matthews_corrcoef(y_test, predictions)
    f1 = f1_score(y_test, predictions, average='macro')
    report = classification_report(y_test, predictions)
    roc_auc = roc_auc_score(y_test, predictProba[:,1])
    accuracy = accuracy_score(y_test, predictions)
    confMatrix = confusion_matrix(y_test, predictions)
    return {
        'report' : report,
        'matthews' : matt,
        'f1' : f1,
        'roc' : roc_auc,
        'accuracy' : accuracy,
        'confusion' : confMatrix
    }

def calc_metrics(X_train, y_train, X_ros , y_ros , X_test, y_test, model) :
    model.fit(X_train, y_train)
    train_error = calc_train_error(X_ros, y_ros, model)
    validation_error = calc_validation_error(X_test, y_test, model)
    return train_error, validation_error
```

```python
K = 5 
kf = KFold(n_splits=K, shuffle=True, random_state=42)
for train_index, test_index in kf.split(data,target) :
    print(train_index)
    print(test_index)
    print(len(train_index))
    print(len(test_index))
    print(data.iloc[train_index].head())
```

```python
train_errors = []
validation_errors = []

for train_index, test_index in kf.split(data, target):
    X_train, X_test = data.iloc[train_index], data.iloc[test_index]
    y_train, y_test = target.iloc[train_index], target.iloc[test_index]
    
    print(len(X_test), (len(X_train) + len(X_test)))
    
    gbr = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1)
    gbr.fit(X_ros,y_ros)
    
    train_error, test_error = calc_metrics(X_train, y_train,  X_ros , y_ros ,X_test, y_test, gbr)
    
    train_errors.append(train_error)
    validation_errors.append(test_error)
```

## 10. Print the Model Evaluation like below from the K Fold 

```python
for i, train_error, val_error in zip(range(1,6,1), train_errors, validation_errors):
    print('Accuracy train '+ str(i) + ' : ' + str(train_error['accuracy']))
    print('Accuracy test '+ str(i) + ' : ' + str(val_error['accuracy']))
    print('ROC AUC train '+ str(i) + ' : ' + str(train_error['roc']))
    print('ROC AUC test '+ str(i) + ' : ' + str(val_error['roc']))
    print('F1 Score train '+ str(i) + ' : ' + str(train_error['f1']))
    print('F1 Score test '+ str(i) + ' : ' + str(val_error['f1']))
    print('Matthews Corr Coef train '+ str(i) + ' : ' + str(train_error['matthews']))
    print('Matthews Corr Coef test '+ str(i) + ' : ' + str(val_error['matthews']))
   
    print('\n')
```

```python
dfModelEval =pd.DataFrame(columns=['Train Accuracy', 'Test Accuracy', 'Train ROC AUC', 'Test ROC AUC', 'Train F1 Score',
                                   'Test F1 Score', 'Train Matthews Corr Coef', 'Test Matthews Corr Coef'], 
                          data=[[train_error['accuracy'], val_error['accuracy'], train_error['roc'], val_error['roc'] , 
                                 train_error['f1'] , val_error['f1'], train_error['matthews'], val_error['matthews'] ]for i, train_error, val_error in zip(range(1,6,1), train_errors, validation_errors)])

dfModelEval
```

```python

print('Accuracy train :'  + str(dfModelEval['Train Accuracy'].mean()))
print('Accuracy test : ' +  str(dfModelEval['Test Accuracy'].mean()))
print('ROC AUC train : ' +  str(dfModelEval['Train ROC AUC'].mean()))
print('ROC AUC test : '  + str(dfModelEval['Test ROC AUC'].mean()))
```
