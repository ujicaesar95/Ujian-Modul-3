## 6. Do Standard Scaler for all variable except the target

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df.drop('Class',axis=1))
scaled_features = scaler.transform(df.drop('Class', axis=1))
df_feat = pd.DataFrame(scaled_features,columns=df.columns[:-1])
df_feat.head()
```

## 7. Do PCA for the features to make it only 10 features remaining

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=10)
pca.fit(df_feat)
dfPCA = pca.transform(df_feat)
dfPCA.shape
```

#### and go make the data dataframe results of the PCA like below

```python
dfPCA = pd.DataFrame(dfPCA, columns=['Feature'+str(i) for i in range(1,11)])
dfPCA.head()
```

## 8. Make target series dan show it value_counts in text and seaborn countplot

## 9. Use K Fold (K=5,shuffle=True,random_state=42), every loop use RandomOverSampler for the training data, and then use GradientBoostingClassifier for the model

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=5, shuffle=True, random_state=42)
```

```python
from sklearn.metrics import classification_report,confusion_matrix, roc_auc_score, accuracy_score, roc_curve,auc, matthews_corrcoef, f1_score
from sklearn.ensemble import GradientBoostingClassifier
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=101)
```

```python
def calc_train_error(X_train, y_train, model):
    predictions = model.predict(X_train)
    predictProba = model.predict_proba(X_train)
    matt = matthews_corrcoef(y_train, predictions)
    f1 = f1_score(y_train, predictions, average='macro')
    report = classification_report(y_train, predictions)
    roc_auc = roc_auc_score(y_train, predictProba[:,1])
    acc = accuracy_score(y_train, predictions)
    confMatrix = confusion_matrix(y_train, predictions)
    return {'report': report, 'matthew': matt, 'f1': f1, 'roc': roc_auc, 'accuracy': acc, 'confusion': confMatrix}

def calc_validation_error(X_test, y_test, model):
    predictions = model.predict(X_test)
    predictProba = model.predict_proba(X_test)
    matt = matthews_corrcoef(y_test, predictions)
    f1 = f1_score(y_test, predictions, average='macro')
    report = classification_report(y_test, predictions)
    roc_auc = roc_auc_score(y_test, predictProba[:,1])
    acc = accuracy_score(y_test, predictions)
    confMatrix = confusion_matrix(y_test, predictions)
    return {'report': report, 'matthew': matt, 'f1': f1, 'roc': roc_auc, 'accuracy': acc, 'confusion': confMatrix}

def calc_metrics(X_train, y_train, X_test, y_test, model):
    model.fit(X_train, y_train)
    train_error = calc_train_error(X_train, y_train, model)
    validation_error = calc_validation_error(X_test, y_test, model)
    return train_error, validation_error
```

```python
train_errors = []
validation_errors = []

for train_index, val_index in kf.split(dfPCA, target):
    X_train, X_val = dfPCA.iloc[train_index], dfPCA.iloc[val_index]
    y_train, y_val = target.iloc[train_index], target.iloc[val_index]
    
    X_train, y_train = ros.fit_sample(X_train,y_train)

    gbc = GradientBoostingClassifier()
    
    train_error, val_error = calc_metrics(X_train, y_train, X_val, y_val, gbc)
    
    train_errors.append(train_error)
    validation_errors.append(val_error)
```

## 10. Print the Model Evaluation like below from the K Fold 

```python
dfEval = []
for tr,val in zip(train_errors, validation_errors):
    dfEval.append([tr['accuracy'], val['accuracy'], tr['roc'], val['roc'], tr['f1'], val['f1'],
                  tr['matthew'], val['matthew']])

dfEval = pd.DataFrame(dfEval, columns=['Train Accuracy','Test Accuracy','Train ROC AUC','Test ROC AUC',
                                       'Train F1 Score','Test F1 Score','Train Matthews Corr Coef',
                                       'Test Matthews Corr Coef'])
dfEval
```

```python
print('AVG Accuracy Train : ' + str(dfEval['Train Accuracy'].mean()))
print('AVG Accuracy Validation : ' + str(dfEval['Test Accuracy'].mean()))
print('AVG ROC AUC Train : ' + str(dfEval['Train ROC AUC'].mean()))
print('AVG ROC AUC Validation : ' + str(dfEval['Test ROC AUC'].mean()))
```
